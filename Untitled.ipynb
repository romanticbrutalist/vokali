```python
(al1.Var=='-').sum()
```

```python
al1_va=al1[(al1['MSX']!= '-')&(al1['A2']!= '-')& (al1['Var'] != '-' )& (al1['HX'] != '-' )]
```

```python
al1_va=al1_va[['MSX','A2','Var','HX','MS']]
```

```python
for i in range(len(al1_va.columns)-1):
    for j in range(len(al1_va)):
        al1_va.iloc[j,i]='.'.join(al1_va.iloc[j,i].split(','))
```

```python
for i in range(len(al1_va.columns)-1):
    al1_va.iloc[:,i] = al1_va.iloc[:,i].astype(float)
```

```python
for i in range(len(al1_va)):
        al1_va.iloc[i,-1] = int(al1_va.iloc[i,-1][0])+int(al1_va.iloc[i,-1][4])
```

```python
for i in range(len(al1_va.columns)-1):
        al1_va.iloc[:,i] = 1/al1_va.iloc[:,i]
```

```python
for i in range(len(al1_va.columns)-1):
     al1_va.iloc[:,i]=al1_va.iloc[:,i]/al1_va.iloc[:,i].max()
```

```python
X=al1_va.iloc[:,:-1].values
Y=al1_va.iloc[:,-1].values
```

```python
#clustering = SpectralClustering(n_clusters=3,assign_labels="discretize",random_state=0).fit(X)
clustering = AffinityPropagation().fit(X[:800])
```
