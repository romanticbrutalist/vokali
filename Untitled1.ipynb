```python
def Probability(rating1, rating2): 
  
    return 1.0 * 1.0 / (1 + 1.0 * math.pow(10, 1.0 * ((rating1) - rating2) / 400)) 
  
  
# Function to calculate Elo rating 
# K is a constant. 
# d determines whether 
# Player A wins or Player B.  
def EloRating(Ra, Rb, K, d): 
   
      
    # To calculate the Winning 
    # Probability of Player B 
    Pb = Probability(Ra, Rb) 
  
    # To calculate the Winning 
    # Probability of Player A 
    Pa = Probability(Rb, Ra) 
  
    # Case -1 When Player A wins 
    # Updating the Elo Ratings 
    if (d == 1) : 
        er=1 - Pa
        Ra = Ra + K * (1 - Pa) 
        Rb = Rb + K * (0 - Pb) 
      
  
    # Case -2 When Player B wins 
    # Updating the Elo Ratings 
    elif (d==0.5) : 
        er=abs(0.5 - Pb)
        Ra = Ra + K * (0.5 - Pa) 
        Rb = Rb + K * (0.5 - Pb) 
      
    else : 
        er=abs(0 - Pa)
        Ra = Ra + K * (0 - Pa) 
        Rb = Rb + K * (1 - Pb) 
      
    return Ra,Rb,(er)
    print("Updated Ratings:-") 
    print("Ra =", round(Ra, 6)," Rb =", round(Rb, 6))

def Probabilityad(rating1, rating2): 
  
    return 1.0 * 1.0 / (1 + 1.0 * math.pow(10, 1.0 * ((rating1) - rating2) / 400)) 


def EloRatingad(Ra, Rb, K, d): 
   
      
    # To calculate the Winning 
    # Probability of Player B 
    Pb = Probabilityad(Ra, Rb) 
  
    # To calculate the Winning 
    # Probability of Player A 
    Pa = Probabilityad(Rb, Ra) 
  
    # Case -1 When Player A wins 
    # Updating the Elo Ratings 
    if (d == 1) : 
        er=1 - Pa
        Ra = Ra + K * (1 - Pa) 
        Rb = Rb + K * (0 - Pb) 
      
  
    # Case -2 When Player B wins 
    # Updating the Elo Ratings 
    er=abs(0 - Pa)
    Ra = Ra + d*K * (0 - Pa) 
    Rb = Rb + d*K * (1 - Pb) 
      
    return Ra,Rb,(er)
    print("Updated Ratings:-") 
    print("Ra =", round(Ra, 6)," Rb =", round(Rb, 6))
```

```python
def optimal_k(matches,epoch,lr):    
    rat_dic={}
    ki=[]
    ei=[[],[],[]]
    #k=[0,0,0]
    for k in tqdm(range(0,epoch,20)):
        
        error=[0,0,0]
        for i in matches.HomeTeam.unique():
            rat_dic[i]={"tr":1000,"ar":1000,"dr":1000,"nm":0}
    
        for i in range(len(matches)):
            rat_dic[matches.iloc[i,2]]["tr"],rat_dic[matches.iloc[i,3]]["tr"],er0=EloRating(rat_dic[matches.iloc[i,2]]["tr"]+k,rat_dic[matches.iloc[i,3]]["tr"],350,1*(matches.iloc[i,4]>matches.iloc[i,5]) or 0.5*(matches.iloc[i,4]==matches.iloc[i,5]))
            rat_dic[matches.iloc[i,2]]["ar"],rat_dic[matches.iloc[i,3]]["dr"],er1=EloRating(rat_dic[matches.iloc[i,2]]["ar"]+k,rat_dic[matches.iloc[i,3]]["dr"],800,1*(matches.iloc[i,4]>0))
            rat_dic[matches.iloc[i,2]]["dr"],rat_dic[matches.iloc[i,3]]["ar"],er2=EloRating(rat_dic[matches.iloc[i,2]]["dr"]+k,rat_dic[matches.iloc[i,3]]["ar"],600,1*(matches.iloc[i,5]==0))
            rat_dic[matches.iloc[i,2]]["nm"]+=1
            rat_dic[matches.iloc[i,3]]["nm"]+=1
            if(i>len(matches)-100):
                error[0]+=er0
                error[1]+=er1
                error[2]+=er2
        print("k",k)
        print("e",error)
        #print("diff",lr*(2/len(last_two))*error[0],lr*(2/len(last_two))*error[1],lr*(2/len(last_two))*error[2])
        #k[0]=k[0]+lr*(2/len(last_two))*error[0]
        #k[1]=k[1]+lr*(2/len(last_two))*error[1]
        #k[2]=k[2]+lr*(2/len(last_two))*error[2]
        ei[0].append(error[0])
        ei[1].append(error[1])
        ei[2].append(error[2])

        ki.append(k)
    return ei,ki
        
    
```
